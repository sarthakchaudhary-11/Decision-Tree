```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df
```

```python
data.target
```

```python
X = df
y = data.target
```

```python
X
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion='entropy')
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12, 10))
tree.plot_tree(classifier, filled=True)
#plt.show() #not to see the entropy details
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
#post pruning >> after making complete decision tree you have to cut it to avoid overfitting
#for small dataset only

#observation
#observe the decision tree plot, you will realise the last split is actually overfitting, diving the data 
#in such as way that each node is 100 percent pure leading to low no of data point, i.e-1 and 3
#so reduce the depth of dt
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion='entropy', max_depth=2)
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12, 10))
tree.plot_tree(classifier, filled=True)
#plt.show() #not to see the entropy details
```

```python
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
#the accuracy is same, earlier it was a overfitting model
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#prepruning using hyperparameter tuning

parameter = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5, 6, 7, 10],
    'max_features':['sqrt', 'log2', 'auto']
}
```

```python
from sklearn.model_selection import GridSearchCV
clf = DecisionTreeClassifier()
model=GridSearchCV(clf, param_grid=parameter, cv = 5, scoring="accuracy", verbose=3)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred = model.best_estimator_.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
